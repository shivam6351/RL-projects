# Q1 : Results from PPO, DDPG and A3C

In this task we wanted to reproduce the results from the MB-MPO paper. since MB-MPO wasnt working on the system. we tried to reproduce the result on PPO and DDPG. firstly in order to run for specefic episodes say 10^5. we would first go to anaconda terminal and type this:
> conda activate rllib
rllib train --run {algo} --env {env} --config {} --stop {"""timesteps_total""": episode}
in the config you can change hyper parameters accordingly. what we did was we ran PPO Ant for various gamma and lr for about 1M eps. and then we plotted the PPO for the one config that gave us the best result.

```python
#parameter selection 
#gamma = 0.99 and lr = 0.0001


import json
import numpy as np
# Read the JSON file
data_ppo_ant_1 = []
with open('ray_results/default/PPO_Ant-v4_a01e7_00000_0_2023-04-26_11-08-51/result.json') as f:
    for obj in f:
        data_ppo_ant_1.append(json.loads(obj))

# Extract all episode rewards
rewards = []
iter = 0
for obj in data_ppo_ant_1:
    for key in obj:
        if(key == "hist_stats"):
            rewards.append(obj[key]["episode_reward"])
            iter+=1

# Compute the average of each reward array
averages_ppo_ant_1 = []

for reward in rewards:
    average_reward = np.sum(reward) / len(reward)
    averages_ppo_ant_1.append(average_reward)

    
print(averages_ppo_ant_1[len(averages_ppo_ant_1) - 1])
```

# Q2: Transfer Learning in RL.

In this Task we had to do transfer learning from one env to another. i trained the Ant-v4 env using the PPO algorithm and used the policies obtained there to learn the halfcheetah env. 

The action space and observation space are diffrent for Ant and halfcheetah env. the observation space for ant is 27 while that of the haldcheetah is 17. The action space of ant is 8 while that of haldfcheetah is 6. so in order to get the action i needed to paas a observation state of 27. so i added 10 "0" values states in obs of halfcheetah. that gave me a 8 length array. i needed to get and action space of 6 so i took the most matching action from both env and ran the action, obtained the next state,reward and wheather its terminal state or not.

i checked for gamma = 0.99,lr = 10^-4 and gamma = 0.95 and lr = 10^-5. and results for the later one were better.

```python
from ray.rllib.algorithms.algorithm import Algorithm
import numpy as np
import gymnasium as gym
from ray.rllib.policy.policy import Policy

policy = Policy.from_checkpoint("ray_results/default/PPO_Ant-v4_bbcff_00000_0_2023-04-26_15-34-29/checkpoint_002500/policies/default_policy/")
# Continue training

env = gym.make("HalfCheetah-v4")
env.gamma = 0.95
env.lr = 0.00001
episode_reward = 0.0
obs = env.reset()[0]
obs = list(obs)
for i in range(10):
    obs.append(0)
obs = np.array(obs)

done = False
iter = 0
epi = []
while iter<5000:
    iter = iter + 1
    action = policy.compute_single_action(obs)
    ac = action[0]
    ac = ac[:6]

    obs, reward, done, _,info = env.step(ac)
    obs = list(obs)
    for i in range(10):
        obs.append(0)
    obs = np.array(obs)
    episode_reward += reward
    epi.append(reward)
print("Episode reward:", episode_reward)
avg_epi = []
sum = 0
iter = 0
for i in epi:
    sum+=i
    iter+=1
    avg_epi.append(sum/iter)
plt.plot(avg_epi)
```
